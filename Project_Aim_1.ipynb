```python
import pandas as pd
import matplotlib.pyplot as plt

#Source data was pulled from the The World Bank website.

df = pd.read_csv('../API_EG.ELC.RNWX.KH_DS2_en_csv_v2_2264150.csv')
```

```python
#Project question one: How has renewable energy generation changed across the globe over time?
```

```python
df.head(10)
```

```python
#using the .iloc method allows you to work with particular rows and columns. The syntax is [row:row, column:column]
#The following code will get rid of the first three rows which are worthless, by creating a new dataframe starting
#from the third row. 
filtered_df = df.iloc[3:, :]
filtered_df.head(10)
```

```python
#Now we need to make the third row the header row, which is the year.
filtered_df.columns = df.iloc[3]
```

```python
filtered_df.head(10)
```

```python
#Now use the drop method to drop the duplicate row by index.
filtered_df = filtered_df.drop(index = 3)
filtered_df
```

```python
#Now lets filter the source data again to include only countries from 'countries_codes_and_coordinates.txt'
#I'm doing this because the source data has way more than just recognized countries. It has disputed territories and
#groups of countries as well, and I only want this data for each independent country. 

with open('../countries_codes_and_coordinates.csv') as f:
    countries = [line.split(',')[2]for line in f]

countries[:5]
```

```python
filtered_df.rename(columns = {"Country Code": "Country_Code"}, 
          inplace = True)
  
```

```python
#We are trying to match strings between the 'countries' list and the 'Country_Code' column in filterd_df.
#To do this, you have to remove the extra quotes so a match will occur.
countries_no_quotes = [s.replace('"','') for s in countries]
countries_no_quotes = [s.strip() for s in countries_no_quotes]
countries_no_quotes[:5]
#The isin() function will look for matches between a dataframe column and a list
filtered_df.Country_Code.isin(countries_no_quotes)
```

```python
#Filter the dataframe again to only include independent countries
countries_only_df = filtered_df[filtered_df.Country_Code.isin(countries_no_quotes)]
countries_only_df
```

```python
#Lets go ahead and reset the indeces again. Best practices.
countries_only_df = countries_only_df.reset_index()
countries_only_df = countries_only_df.drop(columns = 'index')
countries_only_df
```

```python
#Now we can do some analysis.
```

```python
#I need to create a new dataframe with two columns. One is the year, and one is the sum of energy 
#usage for all countries for that year.

years = countries_only_df.columns
print(years)
```

```python
years = years[4:]
```

```python
print(years)
```

```python
#Now that I've got years isolated, I need to isolate the sum of energy generation for each year across the globe. 
#I think the next step would be to use a for loop and iterate over each column (year), taking the sum and adding it 
#to a list.
summed_energy_usage = []
for column in countries_only_df.columns[4:]:
    a = countries_only_df[column].sum()
    summed_energy_usage.append(a)
    
print(summed_energy_usage)
    
```

```python
#Lets make sure both of these lists have the same number of values so we can make a scatterplot.
len(years)
```

```python
len(summed_energy_usage)
```

```python
df_final = pd.DataFrame({'Year': years, 'Total Energy Generation by Renewables Globally (kWh)': summed_energy_usage})
```

```python
df_final.head(10)
```

```python
plt.scatter(df_final['Year'], df_final['Total Energy Generation by Renewables Globally (kWh)'])
plt.xlabel('Year')
plt.ylabel('Total Energy Generation by Renewables Globally (kWh)')
```
