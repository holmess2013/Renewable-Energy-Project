```python
import pandas as pd
import matplotlib.pyplot as plt

#Source data was pulled from the The World Bank website.
```

```python
#Project question 2: What are the top ten countries generating renewable energy?
```

```python
source_df = pd.read_csv('../API_EG.ELC.RNWX.KH_DS2_en_csv_v2_2264150.csv')
source_df.head()
```

```python
#using the .iloc method allows you to work with particular rows and columns. The syntax is [row:row, column:column]
#The following code will get rid of the first three rows which are worthless, by creating a new dataframe starting
#from the third row. 
filtered_df = source_df.iloc[3:,:]
filtered_df.head()
```

```python
#How do you reassign indeces? 
filtered_df = filtered_df.reset_index()
filtered_df = filtered_df.drop(columns = 'index')
filtered_df.head()
```

```python
#Now we need to make the third row the header row, which is the year.
filtered_df.columns = filtered_df.iloc[0]
filtered_df.head()
```

```python
#Drop the duplicate row
filtered_df = filtered_df.drop(index = 0)
filtered_df.head()
```

```python
#Lets go ahead and reset the indeces again.
filtered_df = filtered_df.reset_index()
filtered_df = filtered_df.drop(columns = 'index')
filtered_df.head()
```

```python
#Now lets filter the source data again to include only countries from 'countries_codes_and_coordinates.txt'
#I'm doing this because the source data has way more than just recognized countries. It has disputed territories and
#groups of countries as well, and I only want this data for each independent country. 

with open('../countries_codes_and_coordinates.csv') as f:
    countries = [line.split(',')[2]for line in f]

#The easiest way to match countries is by the three letter code
countries[:5]
```

```python
#Have to remove the space from the column header for use in a later function- the isin() function.
filtered_df.rename(columns = {"Country Code": "Country_Code"}, 
          inplace = True)
filtered_df.head()
```

```python
#We are trying to match strings between the 'countries' list and the 'Country_Code' column in filterd_df.
#To do this, you have to remove the extra quotes so a match will occur.
countries_no_quotes = [s.replace('"','') for s in countries]
countries_no_quotes = [s.strip() for s in countries_no_quotes]
countries_no_quotes[:5]
```

```python
#The isin() function will look for matches between a dataframe column and a list
filtered_df.Country_Code.isin(countries_no_quotes)
```

```python
#Filter the dataframe again to only include independent countries
countries_only_df = filtered_df[filtered_df.Country_Code.isin(countries_no_quotes)]
countries_only_df
```

```python
#Reset the indeces a final time. ALWAYS RESET INDECES AFTER YOU FILTER YOUR DATAFRAME!!!!!!!
countries_only_df = countries_only_df.reset_index()
countries_only_df = countries_only_df.drop(columns = 'index')
countries_only_df
```

```python
#Now we can do some analysis.
```

```python
#I need to sum electricity generation by country, and include it as a new column in the dataframe.
#The axis parameter is very helpful. Basically, if you set axis = 0, it's going to sum across the columns.
#If you give axis = 1, it's going to sum across indeces.
#The next line of code says sum across indeces (the rows) from the 4th column index to the final column index. Then,
#add that column to the dataframe.
countries_only_df['Sum'] = countries_only_df.iloc[:, 4:].sum(axis=1)
countries_only_df
```

```python
#Next step is to verify that those summed values are actually correct.
#I will pick one country to test, Armenia
armenia = countries_only_df.iloc[7, 4:]
armenia

```

```python
#Now lets put it in its own dataframe
df_sum_verify = pd.DataFrame(armenia)
df_sum_verify.head()
```

```python
#The headers are messed up, lets manually fix them
df_sum_verify.columns = ['Electricity Generation (kWh)']
df_sum_verify.head()
```

```python
#Finally, lets clean up the data a little bit and get rid of all the NaN values.
df_sum_verify_cleaned = df_sum_verify.dropna()
df_sum_verify_cleaned
```

```python
#Lets calculate the sum a different way and make sure they match.
summed_energies = 0
for i in df_sum_verify_cleaned['Electricity Generation (kWh)'][:-1]:
    summed_energies = summed_energies + i
print(summed_energies)
```

```python
#They match. Looks like the sum function works correctly!
#Now we just have to go back to filtered_df and sort by 'Sum'
countries_only_df.sort_values(by='Sum', ascending = False)
```

```python
#Finally, we only want the top ten countries.
df_sorted = countries_only_df.sort_values(by='Sum', ascending = False)
df_top_ten = df_sorted.iloc[:10,:]
df_top_ten
```
